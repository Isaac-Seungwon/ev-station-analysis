```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from plotnine import *
import seaborn as sns
```

```python
from google.colab import files
electric_vehicles_registrations = files.upload()
```

```python
import io
electric_vehicles_registrations = pd.read_csv(io.StringIO(electric_vehicles_registrations['integrated data.csv'].decode('cp949')), header = 0, engine = 'python')
```

```python
electric_vehicles_registrations.head()
```

```python
# 그래프를 그릴 도화지 준비하기
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# x축에 년도, y축에 전기차 데이터가 있는 선 그래프 만들기
plt.plot(electric_vehicles_registrations['year'], electric_vehicles_registrations['electric_vehicle'], color='green', marker='o', linestyle='solid')

# 제목 더하기
plt.title("electric vehicles registrationsn Change")

# y축에 레이블 추가하기
plt.ylabel("electric vehicle")
plt.show()
```

```python
# 그래프를 그릴 도화지 준비하기
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# 막대그래프 그리기, 막대 두께 조절
plt.bar(electric_vehicles_registrations['year'], electric_vehicles_registrations['electric_vehicle'], 2.5)

# 제목 더하기
plt.title("electric vehicles registrationsn Change")

# y축에 레이블 추가하기
plt.ylabel("electric vehicle")
plt.show()
```
